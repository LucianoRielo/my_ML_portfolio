### Test set

#### Consideramos que los ingresos medios son un factor importante
Discretizamos los valores, y disminuimos la cantidad de 

```python
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)  # Se divide por 1.5 para reducir la cantidad de categorias
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
housing["income_cat"].hist(bins=100, figsize=(4,3))
```

Dijimos que el ingreso medio de cada hogar es un factor importante para las predicciones, por lo que, si creamos un data set, se desea que este sea representativo. Para ello se realiza un ***stratified sampling***, el cual obtiene asigna con mayor proporcion instancias aleatorias de las categorias más grandes. 

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=1)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]

print(f'train_set has {len(strat_train_set)} instances')
print(f'test_set  has  {len(strat_test_set)} instances')
```

#### Fijarse que la proporción de cada categoria en el data set y en el test set es similar

```python
housing["income_cat"].value_counts() / len(housing)
```

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
for set in (strat_train_set, strat_test_set):  # se borra la columna con categorias
    set.drop(["income_cat"], axis=1, inplace=True)
```

```python
housing = strat_train_set.copy()
```
